```python
#Converting the csv file into dataframe
import pandas as pd
tweet=pd.read_csv('tweets.csv')
tweet
```

```python
 #Cleaning the dataset - Part 1
 import re
 #Converting the tweets into lists
 tweets= tweet['Tweets'].tolist()
 
#Removing the single line character
tweets = [re.sub('\s+', ' ', sent) for sent in tweets]

#Removing the single quote sign
tweets = [re.sub("\'", "", sent) for sent in tweets]
tweets
```

```python
#Cleaning the dataset - Part 2
#Converting Sentences into list of words
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

def s_t_w(j):
    for i in j:
        yield(gensim.utils.simple_preprocess(str(i), deacc=True))  

words = list(s_t_w(tweets))
words
```

```python
#Setting up bigram and trigram
bigram = gensim.models.Phrases(words, min_count=5, threshold=100) 
trigram = gensim.models.Phrases(bigram[words], threshold=100)  

bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
#Cleaning dataset - Part 3
#Removing stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

def remove_stopwords(i):
    return [[word for word in simple_preprocess(str(j)) if word not in stop_words] for j in i]

def make_bigrams(i):
    return [bigram_mod[j] for j in i]

def make_trigrams(i):
    return [trigram_mod[bigram_mod[j]] for j in i]

def lemmatization(i, allowed=['NOUN', 'ADJ', 'VERB', 'ADV']):
    out = []
    for sent in i:
        j = nlp(" ".join(sent)) 
        out.append([token.lemma_ for token in j if token.pos_ in allowed])
    return out

# Removing Stop Words
new_words = remove_stopwords(words)

# Creating Bigrams
words_bigrams = make_bigrams(new_words)

# python3 -m spacy download en
nlp = spacy.load('en', disable=['parser', 'ner'])

# Do lemmatization keeping only noun, adj, vb, adv
cleaned = lemmatization(words_bigrams, allowed=['NOUN', 'ADJ', 'VERB', 'ADV'])

id2word = corpora.Dictionary(cleaned)
texts = cleaned
texts
```

```python
# Calculating Frequency of words
corpus = [id2word.doc2bow(i) for i in texts]
[[(id2word[id], freq) for id, freq in cp] for cp in corpus]


# Building LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=10, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
print(lda_model.print_topics())
```

```python
#Importing and defining dataframe
tweets=pd.read_csv('tweets.csv')
tweets
```

```python
import nltk
nltk.download('wordnet')
```

```python
#Defining the function for data cleaning
#Importing the Libraries
import string
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer

stopwords = set(stopwords.words('english'))
excludepun = set(string.punctuation)
lem = WordNetLemmatizer()

#Data Cleaning Function
def cleaning(text):
    stop_free = ' '.join([word for word in text.lower().split() if word not in stopwords])
    punc_free = ''.join(ch for ch in stop_free if ch not in excludepun)
    normalized = ' '.join([lem.lemmatize(word) for word in punc_free.split()])
    return normalized.split()
```

```python
#More Cleaning
import warnings
warnings.simplefilter('ignore')
tweets['Cleaned_Tweets']=tweets['Tweets'].apply(cleaning)

#Selecting the final dataframe to be used
tweets = tweets[['Tweets','Sentiments','Cleaned_Tweets']]
tweets
```

```python
#Feature Selection
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import svm
from sklearn import naive_bayes
```

```python
#Defining Vector
vector = TfidfVectorizer(ngram_range=(1,2), max_features=1000)
vector.fit(tweets['Sentiments'])

#Calculating x and y values
x = vector.transform(tweets['Sentiments'])
encoder = LabelEncoder()
y = encoder.fit_transform(tweets['Sentiments'])

#Defining the train and test values with 20% of test and 80% train
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.2)
```

```python
# Applying SVM Model
#Defining the model
model_svm = svm.SVC(kernel='linear')
#Applying the model
model_svm.fit(x_train, y_train)
#Calculating the predicted values
y_predicted = model_svm.predict(x_test)
#Calculating the accuracy
print("Accuracy score is {0}".format(accuracy_score(y_test, y_predicted)))
report = classification_report(y_test, y_predicted, output_dict=True)
#Displaying report
report
```

```python
# Applying Naive Bayes Model
#Defining the model
naive_model = naive_bayes.MultinomialNB()
#Applying the model
naive_model.fit(x_train, y_train)
#Calculating the predicted values
predicted_y = naive_model.predict(x_test)
#Calculating the accuracy
print("Accuracy score is {0}".format(accuracy_score(y_test, predicted_y)))
report_naive = classification_report(y_test, predicted_y, output_dict=True)
#Displaying report
report_naive
```
