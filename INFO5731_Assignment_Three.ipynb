```python
# Question 2 - Part 1
#Importing the math library
import math
#Fetching the words and converting it to the list - words in below
words=f['Tweets'].values.tolist()
#Creating the new list by splitting
new_list=set([x for y in words for x in y.split(' ')])
#Creating dataframe for tf_idf
tf_idf=pd.DataFrame(new_list,columns=['token'])

#Creating a function that will calculate the tf*idf values
def func(s,w):
  split=len(s.split(' '))
  value_tf=s.count(w)/split
  value_idf=0
  if(value_tf!=0):
    value_idf=math.log(split)/s.count(w)
  else:
    return 0;
  return value_tf*value_idf

#Defining the counts as 0
counts=0
#Calling the function inside the datafarme we just created
for i in words:
  tf_idf[str(counts)]=tf_idf['token'].apply(lambda x: func(i,x)) #Function Call
  counts+=1
#Final Result DataFrame
tf_idf

#Question 2- Part 1 Complete
```
