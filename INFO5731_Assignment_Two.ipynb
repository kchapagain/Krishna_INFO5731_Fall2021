```python
#Part 1 - Basic feature extraction using text data
#importing the required libraries
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import nltk
from nltk.corpus import *

#setting up the url for the given text file from github
url =  urllib.request.urlopen("https://raw.githubusercontent.com/unt-iialab/info5731-fall2021/main/assignments/01-05-1%20%20Adams%20v%20Tanner.txt")
soup = BeautifulSoup(url)

#Text file as output in string format
output = soup.get_text().replace("\n", "").strip()

#Below codes give the count of requirements of question

Sentence_Count = len(output.split("."))


Words_Count = len(output.replace("."," ").split())


Characters_Count = len(output)


Average_Word_Length = round(sum(len(word) for word in output.split()) / len(output.split()))


Stop_Words_List = []
Stop_Words = stopwords.words('english')

for i in output.split():
  if i in Stop_Words:
    Stop_Words_List.append(i)
    
Total_Stopwords_Count = len(Stop_Words_List)


Special_Characters = '[@+-_!#=$%;^&*(|)<>?/\|}{~:]'
Special_Characters_List = []
for i in output:
  if i in Special_Characters:
    Special_Characters_List.append(i)

Special_Characters_Count = len(Special_Characters_List)


Numerics_Count = 0
for i in output:
    if i.isdigit():
        Numerics_Count += 1
Total_Numerics_Count = Numerics_Count


# For total number of upercase words.
uppercase_words_count = 0
for i in output:
    if i.isupper():
        uppercase_words_count += 1
Uppercase_Words_Count = uppercase_words_count

#printing the output in clean format

print('Sentence_Count           :  ' ,Sentence_Count)
print('Words_Count              :  ' ,Words_Count)
print('Characters_Count         :  ' ,Characters_Count)
print('Average_Word_Length      :  ' ,Average_Word_Length)
print('Total_Stopwords_Count    :  ' ,Total_Stopwords_Count)
print('Special_Characters_Count :  ' ,Special_Characters_Count)
print('Total_Numerics_Count     :  ' ,Total_Numerics_Count)
print('Uppercase_Words_Count    :  ' ,Uppercase_Words_Count)
```

```python
#downloading the required libraries
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
#Part 2- Basic Text Pre-processing of text data

#converting the output into dataframe
Cleaned_File = pd.DataFrame()
Cleaned_File['Sentences'] = output.split('\r')

#Below codes transforms the sentences step by step

#Converting all to lowercase letters
Cleaned_File['Sentences'] = Cleaned_File['Sentences'].apply(lambda x: " ".join(x.lower() for x in x.split()))
Cleaned_File = Cleaned_File[Cleaned_File['Sentences'].str.len()>=20]

#Removing the Punctuations
Cleaned_File['Sentences'] = Cleaned_File['Sentences'].str.replace('[^\w\s]','')

#Removing the Stop Words
Cleaned_File['Sentences'] = Cleaned_File['Sentences'].apply(lambda x: " ".join(x for x in x.split() if x not in Stop_Words_List))

#Removing the common words
common = pd.Series(' '.join(Cleaned_File['Sentences']).split()).value_counts()[:10]

Cleaned_File['Sentences'] = Cleaned_File['Sentences'].apply(lambda x: " ".join(x for x in x.split() if x not in common))

#Removing the rare words
rare = pd.Series(' '.join(Cleaned_File['Sentences']).split()).value_counts()[-10:]
Cleaned_File['Sentences'] = Cleaned_File['Sentences'].apply(lambda x: " ".join(x for x in x.split() if x not in rare))

#Spelling Correction, Tokenization, Tokenization,Stemming and Lemmatization
from textblob import TextBlob
Cleaned_File['Sentences'][:5].apply(lambda x: str(TextBlob(x).correct()))

TextBlob(Cleaned_File['Sentences'][1]).words

from nltk.stem import PorterStemmer
st = PorterStemmer()
Cleaned_File['Sentences'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

from textblob import Word
Cleaned_File['Sentences'] = Cleaned_File['Sentences'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

print(Cleaned_File)
```

```python
#Part 3 - Save all the clean sentences to a csv file (one column, each raw is a sentence) after finishing all the steps above.

#Converting the Dataframe to csv file
Cleaned_File_csv = Cleaned_File.to_csv(index=False)
Cleaned_File_csv
```

```python
#Part 4 - Advance Text Processing

term_frequency = (Cleaned_File['Sentences'][2:3]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
print(term_frequency)

One_Gram = TextBlob(Cleaned_File['Sentences'][1]).ngrams(1)
Two_Gram = TextBlob(Cleaned_File['Sentences'][1]).ngrams(2)
Three_Gram = TextBlob(Cleaned_File['Sentences'][1]).ngrams(3)
print(One_Gram)
print(Two_Gram)
print(Three_Gram)
```

```python
#Part 1 - Parts of Speech (POS) Tagging
Sentences = Cleaned_File.Sentences.tolist()

#Importing the required libraries
import spacy
nlp = spacy.load('en_core_web_sm') 

#initializing the lists
nouns = []
adjectives = []
verbs = []
adverbs = []

#Loop for counting nous, adjectives,verbs and adverbs
for words in Sentences: 
  word = nlp(words)
  for token in word:
    if token.pos_ == 'NOUN': 
      nouns.append(token)
    if token.pos_ == 'ADJ': 
      adjectives.append(token)
    if token.pos_ == "VERB": 
      verbs.append(token)
    if token.pos_ == "ADV": 
      adverbs.append(token)

#Output as count of each items asked by question
print("Nouns_Count     : ", len(nouns))
print("Verbs_Count     : ", len(verbs))
print("Adjectives_Count: ", len(adjectives))
print("Adverbs_Count   : ", len(adverbs))
```

```python
#Part 2 - Dependency Parsing Parsing
#importing the library for dependency parsing
from spacy import displacy

for words in Sentences: 
  text = nlp(words)
  displacy.render(text, style='dep', jupyter=True)
```

```python
#Part 3 - Named Entity Recognition

#Initializing the empty list for entity
entity_list = []
for words in Sentences:  
  doc = nlp(words)
  for ent in doc.ents:
     entity_list.append(ent.text)

#Counting the entities and returning the count
result = sorted([(x, entity_list.count(x)) for x in entity_list], key=lambda y: y[1]) 
for i in result:
    print ('{} {}'.format(i[0], i[1]))
```
